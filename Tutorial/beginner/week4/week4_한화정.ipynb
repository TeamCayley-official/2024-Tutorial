```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
plt.rc('font',family='Malgun Gothic')
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
df = pd.read_csv("김영삼도서관 장서 대출목록 (2024년 04월).csv", low_memory=False)
def assign_subject(row):
    if 0 <= row['주제분류번호'] < 100:
        return "총류"
    elif 100 <= row['주제분류번호'] < 200:
        return "철학"
    elif 200 <= row['주제분류번호'] < 300:
        return "종교"
    elif 300 <= row['주제분류번호'] < 400:
        return "사회과학"
    elif 400 <= row['주제분류번호'] < 500:
        return "자연과학"
    elif 500 <= row['주제분류번호'] < 600:
        return "기술과학"
    elif 600 <= row['주제분류번호'] < 700:
        return "예술"
    elif 700 <= row['주제분류번호'] < 800:
        return "언어"
    elif 800 <= row['주제분류번호'] < 900:
        return "문학"
    else:
        return "역사"
df["주제"] = df.apply(assign_subject, axis = 1)
total_general_work = df[df["주제"] == "총류"]["대출건수"].sum()
total_philosophy = df[df["주제"] == "철학"]["대출건수"].sum()
total_religion = df[df["주제"] == "종교"]["대출건수"].sum()
total_social_science = df[df["주제"] == "사회과학"]["대출건수"].sum()
total_natural_science = df[df["주제"] == "자연과학"]["대출건수"].sum()
total_technical_science = df[df["주제"] == "기술과학"]["대출건수"].sum()
total_art = df[df["주제"] == "예술"]["대출건수"].sum()
total_language = df[df["주제"] == "언어"]["대출건수"].sum()
total_literature = df[df["주제"] == "문학"]["대출건수"].sum()
total_history = df[df["주제"] == "역사"]["대출건수"].sum()
data = pd.Series({
    "총류": total_general_work,
    "철학": total_philosophy,
    "종교": total_religion,
    "사회과학": total_social_science,
    "자연과학": total_natural_science,
    "기술과학": total_technical_science,
    "예술": total_art, 
    "언어": total_language,
    "문학": total_literature,
    "역사": total_history
})
data = data.sort_values(ascending=False)
data.plot.bar(rot=0, figsize = (10,5), title = "주제별 대출건수")
```
