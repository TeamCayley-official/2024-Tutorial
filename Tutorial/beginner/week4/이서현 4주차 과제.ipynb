```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_excel("도서관정보나루_참여도서관목록.xlsx")
df.shape
```

```python
df[["경도", "위도"]].plot.scatter(x="경도", y="위도", figsize=(8, 7), grid=True)
```

```python
plt.figure(figsize=(9, 8))
sns.scatterplot(data=df, x="경도", y="위도")
```
