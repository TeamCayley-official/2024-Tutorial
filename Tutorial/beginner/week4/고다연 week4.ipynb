```python
# 주제분류기호를 사용하여 김영삼도서관에서 가장 많이 분류된 주제 5개 출력하기

import pandas as pd 
from collections import Counter
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
#dtype: float

df = pd.read_excel("C:/Users/고다연/Downloads/week4/김영삼도서관 장서 대출목록 (2024년 04월).xlsx")

df['첫숫자'] = df['주제분류번호'].astype(str).str.split('.').str[0].str[0]


first_digit_counts = df['첫숫자'].value_counts(normalize=True)


top_5_digits = first_digit_counts.head(5)

top_5_digits.plot.barh(color='salmon')

# 4, 1, 5, 3, 8 말고 자연과학, 철학, 기술과학, 문학으로 뜨게 하는 방법? 여기서 seaborn 코드로 시각화하는 방법은?
```
