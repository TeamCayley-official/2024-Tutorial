```python
#7-4 2.
import random

def lotto(): #로또 1회당 로또 번호
    sixnum=[]
    while len(sixnum) <= 5:
        number = random.randint(1, 45)
        if number not in sixnum:
            sixnum.append(number)
    return sixnum

def lotto_stat():
    results = {}
    for i in range(1,46):
        results[i] = 0 #딕셔너리 생성, key값은 1~45, values값의 초기값은 0  
    for i in range(1, 1001): #1회당 로또 번호를 1000번 하기
        numbers = lotto() #lotto 클래스를 객체로 만들기, sixnum인 [n1, n2, n3, n4, n5, n6]이 할당됨
        print("%d회: " % i, end="") #몇 회인지 출력
        for j in numbers: #n회차의 sixnum에서 차례대로 n1, n2 ... 출력
            print("%d " % j, end="")
        print("\n") #한 칸 내려가기
        for k in numbers: #숫자 개수 세기
            results[k] = results[k] + 1 #k=n1이면 results딕셔너리에서 key값이 n1인 value값에 1 더해주기
    print("\n")
    for i in results.keys():
        print("%s : %d회" % (i,results[i]))
        
    sorted_results = sorted(results.items(), key=lambda x: x[1], reverse=True) # 결과를 값이 큰 순서대로 정렬
    selected_numbers = [num for num, count in sorted_results[:6]] # 상위 6개의 숫자 선택
    print("이 주의 로또 번호 :", end=" ")
    for num in selected_numbers:
        print(num, end=" ")
        
lotto_stat()
```
