```python
df2 = pd.read_csv('1923 agereadbook.csv')
print(df2.head())
```

```python
df2.rename(columns={'영유아.1': '영유아',
                   '유아.1': '유아',
                   '초등.1': '초등',
                   '청소년.1': '청소년',
                   '20대.1': '20대',
                   '30대.1': '30대',
                   '40대.1': '40대',
                   '50대.1': '50대',
                   '60대 이상.1': '60대',
                   '미상.1': '미상'}, inplace=True)

print(df2.head())
```

```python
df2.rename(columns={'연도/지역': '연도',
                   'Unnamed: 1': '지역'}, inplace=True)

df_selected = df2[['연도', '지역', '20대']]

print(df_selected.head())
```

```python
df2['long'] = 0.0
df2['lat'] = 0.0

df2.loc[df2['지역'] == '서울특별시', ['long', 'lat']] = [126.9780, 37.5665]
df2.loc[df2['지역'] == '경기도', ['long', 'lat']] = [127.5183, 37.4138]
df2.loc[df2['지역'] == '부산광역시', ['long', 'lat']] = [129.0756, 35.1796]
df2.loc[df2['지역'] == '경상남도', ['long', 'lat']] = [128.6922, 35.2385]
df2.loc[df2['지역'] == '충청남도', ['long', 'lat']] = [126.7757, 36.6372]

df_selected = df2[['연도', '지역', 'long', 'lat', '20대']]

print(df_selected.head())
```

```python
df_selected.columns.values[4] = '20대 독서량'
df_selected.columns.values[5] = '20대 독서율'
df3 = df_selected[['연도', '지역', 'long', 'lat', '20대 독서량', '20대 독서율']]
df3 = df3.drop(0)
print(df3)
```

```python
df3.loc[1:5, '연도'] = 2019
df3.loc[6:10, '연도'] = 2020
df3.loc[11:15, '연도'] = 2021
df3.loc[16:20, '연도'] = 2022
df3.loc[21:25, '연도'] = 2023

print(df3)
```
